# Main objective of the project is to find the relation of horsepower and MPG of the car
>* Horsepower is a measure of the power of an engine. It is often used to compare the performance of different cars, especially their speed and acceleration. The higher the horsepower, the more work the engine can do in a given time.
>* MPG stands for miles per gallon and is used to show how far your car can travel for every gallon (or 4.55 litres) of fuel it uses. The higher the MPG, the more fuel-efficient the car is. MPG is often used to compare the performance and environmental impact of different cars.
>* The relationship between mpg and horsepower is negative. This means that the more horsepower an automobile has, the less fuel-efficient it is. Now, let us verify it.

### now we use the mean manupulation method so that we can replace null values with mean value
#### Mean manupulation method refers to the method in which we replace the null value with mean value 

```python
#inspecting the null values in the dataset
#print(df.isnull().values.sum())
print(df.isnull().values.sum())

#replace the ' 's with NaN
df = df.replace(" ",np.NaN)
#df=df.replace(" ",np.NaN)
#impute the missing values with mean imputation
df=df.fillna(df.mean())
#count the numbers again for verification
print(df.isnull().values.sum())
```

```python
#seaborn histogram
sns.displot(data=df, x="mpg", kind="hist")
#add labels
plt.title('MGP VS COUNT')
plt.xlabel('MPG')
```

# Interpretation

>* Here MPG of the car data is shown in x axis whereas the count is shown in y axis. 
>* From the surveyed data most of the car have MPG around 10 to 15 and only few cars have MPG around 30 to 35.  

```python
#seaborn histogram
sns.displot(data=df, x="horsepower", kind="kde")
#add labels
plt.title('HORSEPOWER VS COUNT')
plt.xlabel('HORSEPOWER')
```

# Interpretation

>* Here horsepower of the car data is shown in x axis whereas the count is shown in y axis. 
>* From the surveyed data, most of the car have horsepower around 75 to 100 and only few cars have horsepower around 175 to 200.  

```python
sns.barplot(data=df, x="mpg", y="horsepower")
plt.title('Horsepower vs mpg ')
plt.show()
```

# Intrepretation
>* The given chart shows the relationship between horsepower and mpg of the car
>* Cars which has low mpg give the highest horsepower whereas the cars having highest mpg give lowest horsepower

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(16, 12))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features', size = 15)
colormap = sns.diverging_palette(10, 220, as_cmap = True)
sns.heatmap(df.corr(),
            cmap = colormap,
            square = True,
            annot = True,
            linewidths=0.1,vmax=1.0, linecolor='white',
            annot_kws={'fontsize':20 })
plt.show()
```
